```python
#Bcsf19m531
#Cs-Morning
#Assignment 03

#-------------------------------8-Queen Genetic Algorithm Problem Solution------------------------------------#

import random

#List is taken to store solutions and a 8x8 chessboard is initialized

def view(li, index):
    print()
    print(f"Solution number {index + 1}: ", end='')
    print(li)
    print()

    for i in range(8):
        x = li[i] - 1
        for j in range(8):
            if j == x:
                print('[Q]', end='')
            else:
                print('[ ]', end='')
        print()
    
    print()

#Hurustic Function is defined on the base of which solutions will bwe decided which one is better

def Huristic_Func(instance):
    huristic = []
    for i in range(len(instance)):
        j = i - 1
        huristic.append(0)
        while j >= 0:
            if instance[i] == instance[j] or (abs(instance[i] - instance[j]) == abs(i - j)):
                huristic[i] += 1
            j -= 1
        j = i + 1
        while j < len(instance):
            if instance[i] == instance[j] or (abs(instance[i] - instance[j]) == abs(i - j)):
                huristic[i] += 1
            j += 1
    return huristic

#Fitness Function for solutions scans for solutions discrepencies i.e whether queens are clashing or not

def Fitness_Func(instance):
    clashes = 0
    for i in range(len(instance) - 1):
        for j in range(i + 1, len(instance)):
            if instance[i] == instance[j]:
                clashes += 1
    for i in range(len(instance) - 1):
        for j in range(i + 1, len(instance)):
            if abs(instance[j] - instance[i]) == abs(j - i):
                clashes += 1
    return 28 - clashes

#Childs are initialized and then crossover is applied in separated functions

def make_child(instance1, instance2, crossOver):
    newInstance = []
    for i in range(crossOver):
        newInstance.append(instance1[random.randint(0, 7)])
    for i in range(crossOver, 8):
        newInstance.append(instance2[random.randint(0, 7)])
    return newInstance

def revert_child(co):
    global parent1, parent2, child1, child2, crossover
    crossover = co
    child1 = make_child(parent1, parent2, crossover)
    child2 = make_child(parent2, parent1, crossover)

def Mutation(li):
    global crossover, parent1, parent2
    newchange = -1
    while newchange != 0:
        newchange = 0
        tmpli = li
        getHur = Huristic_Func(tmpli)
        index = getHur.index(max(getHur))
        maxFitness = Fitness_Func(tmpli)
        for i in range(1, 9):
            tmpli[index] = i
            if Fitness_Func(tmpli) > maxFitness:
                maxFitness = Fitness_Func(tmpli)
                newchange = i
            tmpli = li
        if newchange == 0:
            for i in range(len(li) - 1):
                for j in range(i + 1, len(li)):
                    if li[i] == li[j]:
                        li[j] = random.randint(1, 8)
        else:
            li[index] = newchange

#Main function where parents are randomly initialized and fitness and huturistic functions are called

if __name__ == "__main__":
    Total_Sol = int(input())
    
    solutions = []
    crossover = 4
    while len(solutions) < Total_Sol:
        parent1 = []
        parent2 = []
        for i in range(8):
            parent1.append(random.randint(1, 8))
            parent2.append(random.randint(1, 8))
        fitnessParent1 = Fitness_Func(parent1)
        fitnessParent2 = Fitness_Func(parent2)
        while fitnessParent1 != 28 and fitnessParent2 != 28:
            revert_child(crossover)
            Mutation(child1)
            Mutation(child2)
            fitnessParent1 = Fitness_Func(child1)
            fitnessParent2 = Fitness_Func(child2)
            parent1 = child1
            parent2 = child2
            
        if Fitness_Func(parent1) == 28:
            if parent1 not in solutions:
                solutions.append(parent1)
        else:
            if parent2 not in solutions:
                solutions.append(parent2)
   

    print("___________________ Solutions __________________")
    print(f"The number of solutions Entered: {Total_Sol}")

    for i in range(len(solutions)):
        view(solutions[i], i)

    print("________________________________________________")
```
